```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Total runtime values collected during tests with k = 2 for datasets of different dimensions 2000, 3000, 6000
df_tot = pd.DataFrame({'k':[2,2,2],'dataset_size':[2000,4000,6000], 'PF':[492.768462,1183.796319,2240.183541], 'PT':[101.733864,196.920613,407.085226], 'IT':[113.856586,233.327358,402.761416]})
```

```python
#Runtime per tuple with k = 2 for the different size of datasets
df_run = pd.DataFrame({'k':[2,2,2], 'dataset_size':[2000, 4000, 6000], 'PF':[0.246384231, 0.29594907975, 0.3733639235], 'PT':[0.050866932, 0.04923015325, 0.06784753766666667], 'IT':[0.056928293, 0.0583318395, 0.06712690266666667] })
```

```python
#Total runtime values collected during tests with k = 3 for the different size of databases
df2_tot = pd.DataFrame({'k':[3,3,3],'dataset_size':[2000,4000,6000], 'PF':[442.490893, 1185.385416,2364.608899], 'PT':[84.166647,198.220066,345.982299], 'IT':[96.66088,239.135602,406.05608]})
```

```python
#Runtime per tuple with k = 3 for different size of databases
df2_run = pd.DataFrame({'k':[3,3,3], 'dataset_size':[2000, 4000, 6000], 'PF':[0.22124544650000003, 0.29634635400000003, 0.39410148316666666], 'PT':[0.0420833235, 0.0495550165, 0.057663716500000003], 'IT':[0.04833044, 0.0597839005, 0.06767601333333334] })
```

```python
#plot image function 
def print_graphic(df, flag, k):
    labels = ['P-F', 'P-T', 'I-T']
    index = df.index
        
    x = np.arange(len(labels))  # the label locations
    width = 0.2  # the width of the bars

    fig, ax = plt.subplots()
    ax.set_ylabel('Total Runtime(s)') if flag else ax.set_ylabel('Runtime(s) per tuple')
    ax.set_xlabel('Configuration')
    ax.set_title('K = ' + k)
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
        
    i = width * int(len(df)/2) * (-1)  # to format the distance bewteen dimensions
    legend = []
    for indx in index:
        dataset_size = df.dataset_size[indx]
        PF_tot = df.PF[indx]
        PT_tot = df.PT[indx]
        IT_tot = df.IT[indx]
        times = [PF_tot, PT_tot, IT_tot]
        legend.append(str(dataset_size)+' tuples')
        ax.bar(x + i, times, width)
        i+=width
    
    ax.legend(legend, loc='best')
    fig.tight_layout()
    plt.show()
```

```python
print_graphic(df_tot, True, '2')
```

```python
print_graphic(df_run, False, '2')
```

```python
print_graphic(df2_tot, True, '3')
```

```python
print_graphic(df2_run, False, '3')
```
